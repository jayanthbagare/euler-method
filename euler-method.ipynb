```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
exp = math.e
```

```python
#Consider a simple function e where e**x is calculated
def e(x):
    return exp**x
```

```python
x = np.arange(0,10,1.0)
y = [e(x) for x in x]
```

```python
#Now let us implement the Euler Method
def euler_method(step_size,num_terms,start_point):
    h = step_size
    n = num_terms
    xprev = start_point
    yprev = exp**xprev
    ex = []
    ey = []

    for i in range(n-1):
        y = yprev + h*(exp**xprev)
        x = xprev + h
        ex.append(x)
        ey.append(y)
        yprev = y
        xprev = x
    return ex,ey
```

```python
x1,y1 = euler_method(1,10,1)
x2,y2 = euler_method(0.1,80,1)
x3,y3 = euler_method(0.01,800,1)
```

```python
plt.plot(x,y,linestyle='dotted')
plt.plot(x1,y1,linestyle='dotted')
plt.plot(x2,y2,linestyle='dotted')
plt.plot(x3,y3,linestyle='dotted')
plt.show()
```
